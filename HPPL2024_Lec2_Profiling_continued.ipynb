```python
%%writefile test_scalene.py

import numpy as np
from math import sqrt

def f(n=1000):
    a = np.eye(n)
    b = a @ a

def g(n=1000):
    a = [0.5] * n
    for i,elem in enumerate(a):
        a[i] = sqrt(elem)
    return a

def io_fun(filename='test.dat',n=1000):
    a = np.eye(n)
    np.savetxt(filename, a)

if __name__=='__main__':
    f(n=5000)
    g(n=10000000)
    io_fun(n=5000)
    print('Done')
```

Домашняя задача 1. Фракталы и множество Жулиа
-----

z = x + 1j * y

Исследовать посследовательность:

$$z_{n+1}=z_{n}^2+C$$

Глобальный вопрос: при заданном $C$, какие начальные значения $z$ будут приводить к сходящемуся ряду?

Как строится картинка фрактала:
* выбираем $C$
* создаем двумерную область, где по оси $x$ будет откладываться вещественная часть $z_0$, а по оси $y$ - мнимая часть $z_0$
* задаем максимальное количество итераций, которое мы продолжаем исследовать ряд
* если при каком-то $n$ модуль $z$ превысит какой-то заданный лимит - возвращаем это $n$, иначе возвращаем максимальное количество итераций
* делая это для $Re(z_0)$ от (-2, 2) и $Im(z_0)$ от (-2, 2) ставим цвет равный $n$ в пикселе, соответствующем $z_0$

```python
import numpy as np
import matplotlib.pyplot as plt
# Initial parameters

c = 0.3 + 0.9j

maxiters = 50

z = 0.-0.9j
abs_z = np.zeros(maxiters)

for n in range(maxiters):
  z = z**2+c
  if(np.abs(z)>2): break
  abs_z[n] = np.abs(z)

plt.plot(abs_z)
```
