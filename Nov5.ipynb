```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ y(t) = a_0\,\sin(\omega_L t)$$

Dimensionless time: $t' = \omega_L t$

$$ y(t) = a_0\,\sin(t)$$

In this units 1 period is equal to $2\pi$

```python
Nsteps = 10000
Nsteps = 2**14
Nsteps = Nsteps - 1

t = np.linspace(-30*2*np.pi, 30*2*np.pi, Nsteps)
y = np.sin(t)

plt.plot(t/2/np.pi, y)
plt.xlim(0,2)
```

```python
tau = 5 * 2*np.pi
y = np.exp(-t**2/2/tau**2)*np.sin(t+0*0.004*t**2)

y += 0.5* np.exp(-t**2/2/tau**2)*np.sin(2*t)
y += 0.33* np.exp(-(t-10*2*np.pi)**2/2/tau**2)*np.sin(3*t)



plt.plot(t/2/np.pi, y)
#plt.xlim(0,2)


# use library: sounddevice
# listen to pulses
```

```python
spec = np.fft.fft(y)
w = np.fft.fftfreq(Nsteps, d=t[1]-t[0]) * 2*np.pi   # convenient

dt = t[1]-t[0]
dw = 1 / Nsteps / dt
w_ny = 1 / 2 / dt

my_w = np.linspace(-w_ny, w_ny, Nsteps)

plt.plot(my_w, np.abs(spec))
```

```python
plt.plot(w, np.abs(spec), '.-')
plt.xlim(0, 4)
```
